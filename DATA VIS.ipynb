```python
import numpy as np
```

```python
x=np.arange(10)
y=x**2
y2=2*x+3
```

```python
plt.plot(x,y2)
plt.show()
```

```python
themes=plt.style.available
print(themes)
```

```python
plt.style.use("seaborn")
```

```python
plt.plot(x,y,color='red',label="APPLE",marker='o')
plt.plot(x,y2,color='blue',label="GUAVA",linestyle="dashed")
plt.xlabel("TIME")
plt.ylabel("PRICE")
plt.title("BADIYA")
plt.legend()
plt.show()
```

```python
#SCATTER PLOTS
```

```python
plt.figure(figsize=(4,3))
plt.scatter(x,y,color='red',label="Apple",marker="^")
plt.scatter(x,y2,color="green",label="Kiwi",linestyle="dashed")
plt.xlabel("Time")
plt.ylabel("Price")
plt.title("Prices of fruits over time")
plt.legend()
plt.show()
```

```python
plt.style.use("dark_background")
x_cord=np.arange(3)*2
plt.bar(x_cord,[10,20,15],width=0.5,label="Current Year",tick_label=["A","B","C"],color="teal")
plt.bar(x_cord+0.5,[20,10,12],width=0.5,label="Next Year",color="orange")
plt.ylim(0,30)
plt.legend()
plt.show()
```

```python
#Histogram
```

```python
xsn=np.random.randn(100)
```

```python
print(xsn)
```

```python
sigma=8
u=70
X=np.round(xsn*sigma+u)
X2=np.round(xsn*5+40)
print(X)
```

```python

plt.style.use("seaborn")
plt.hist(X,alpha=0.8)
plt.hist(X2,alpha=0.9)
plt.xlabel("Marks Range")
plt.ylabel("Freq Count")
plt.show()
```
